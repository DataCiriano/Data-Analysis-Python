# STEP 2: Data Preparation

```python
#Creating a subDataFrame seleccting only the right columns

df = df[['coaster_name', 
    #'Length', 'Speed', 
    'Location', 'Status', 
    #'Opening date',
    #'Type', 
    'Manufacturer', 
    #'Height restriction', 'Model', 'Height',
    #'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
    #'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
    #'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
    #'Track layout', 'Fastrack available', 'Soft opening date.1',
    #'Closing date', 'Opened', 'Replaced by', 'Website',
    #'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
    #'Single rider line available', 'Restraint Style',
    #'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 
    'latitude', 'longitude', 'Type_Main',
       'opening_date_clean', 
    #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', 
    #'height_value', 'height_unit', 
    'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
#Example of Dropping columns in the DataFrame
# df.drop(['Status'], axis=1) axis=1 for dropping the entire column not just a single row
```

```python
df.shape
```

```python
#Change type from str to date
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
```

```python
#Rename columns 

df = df.rename(columns={'coaster_name':'Coaster_Name',
                  'year_introduced':'Year_Introduced',
                  'latitude':'Latitude',
                  'longitude':'Longitude',
                  'opening_date_clean':'Opening_Date',
                  'speed_mph':'Speed(mph)',
                  'height_ft':'Height(ft)',
                  'Inversions_clean':'Inversions',
                  'Gforce_clean':'Gforce'})
```

```python
#Identify columns with null values
df.isna().sum()
```

```python
#Show duplicated values (None)

df.loc[df.duplicated()]
```

```python
#Look for duplicated in a column
df.loc[df.duplicated(subset=['Coaster_Name'])]

#Seems that some rows have duplicated Coaster_
```

```python
#Show an example of rows with the same Coaster_Name
df.query('Coaster_Name == "Crystal Beach Cyclone"')

#This two rows looks identical but the data of Year_Introduced is not
```

```python
#Show anly the rows with not duplicated Coaster_name, Location and Opening_Date
df = df.loc[~df.duplicated(subset=['Coaster_Name','Location','Opening_Date'])].reset_index(drop=True).copy() 
# drop=True a new index is set for the DataFrame, drop=False the original index is mantains as a new column and a new index is generated
df.head()
```

# STEP 3: Feature Understanding
   #### (Univariable analysis)
* Plotting Feature Distribution: Histogram, KDE, Boxplot

```python
#Count how many times a value ocurs in a serie (column)
df['Year_Introduced'].value_counts()
```

```python
#Plot 10 first the values in a bar char
ax = df['Year_Introduced'].value_counts().head(10).plot(kind='bar',title='Top 10 Years Coasters Introduced')
ax.set_xlabel('Year Introduced')
ax.set_label('Count')
```

```python
#Plot speed Coasters distibutions with an histogram
ax = df['Speed(mph)'].plot(kind='hist', bins=20, title='Coaster Speed (mph)')
ax.set_xlabel('Speed (mph)')
```

```python
#Plot speed Coasters distibutions with an Kernel density estimation (KDE)
ax = df['Speed(mph)'].plot(kind='kde',title='Coaster Speed (mph)')
ax.set_xlabel('Speed (mph)')
```

# STEP4: Feature Relationships
* Scaterplot
* Heatmap Correlation
* Pairplot
* Groupby comparisons

```python
#Scatter plot with matplotlib
df.plot(kind='scatter', x='Speed(mph)', y='Height(ft)' ,title='Coaster Speed vs. Height')
plt.show()
```

```python
#Scatter plot with Seaborn
sns.scatterplot(x='Speed(mph)', y='Height(ft)', data=df, hue='Year_Introduced')
```

```python
#Comparing multiple features with pairplot
sns.pairplot(df, vars=['Year_Introduced','Speed(mph)','Height(ft)','Inversions','Gforce'],hue='Type_Main')
plt.show()
```

```python
#Correlation between diferent features
df_corr = df[['Year_Introduced','Speed(mph)','Height(ft)','Inversions','Gforce']].dropna().corr()
#.corr() this funstion shows the correlation between diferents features
```

```python
#Plot the calculated correlation in a heatmap with Seaborn
sns.heatmap(df_corr,annot=True)
```

# STEP 5: Ask a question about the data
* What are the locations with the fastest roller coasters?

```python
df.head()
```

```python
df['Location'].value_counts()
#As we can see 'Other' is not a location so we must get rid of it
```

```python
#Get all Locations diferent than 'Other' and group then together, show Speed mean and only show the locations with 10 at least coasters. Last sort it by the mean asc
ax = df.query('Location != "Other"').groupby('Location')['Speed(mph)'].agg(['mean','count']).query('count > 10').sort_values('mean')['mean'] \
.plot(kind='barh', figsize=(12,5), title='Average Coaster Speed by Location') #Plot results in a horizontal bar char
ax.set_xlabel('Average Coaster Speed')
plt.show()
```
